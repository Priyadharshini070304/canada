```python

import pandas as pd
canadadf=pd.read_csv('canada_per_capita_income.csv')
print(canadadf)
```

```python
x = canadadf[['year']]
y = canadadf[['per capita income (US$)']]
print(x)
print(y)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
ans=model.predict([[2020]])
print(ans)
```

```python
model.score(x,y)
```
